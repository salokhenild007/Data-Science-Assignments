## 1.Data Cleaning and Preparation:

```python
#A) Load the dataset into a DataFrame or equivalent data structure.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('Cardiotocographic.csv')
data.head()
```

```python
data
```

```python
#here NSP is 1 and 2 so
# Replace 1 with 'Yes' and 2 with 'No'
data['NSP'] = data['NSP'].replace({1: 'Yes', 2: 'No'})
```

```python
data
```

```python
#B) Handle missing values appropriately (e.g., imputation, deletion).
data.shape
```

```python
data.columns
```

```python
data.info()
```

```python
# since it has missing values
# Replace missing values with the mean (or median) for each numerical column
data['LB'].fillna(data['LB'].mean(), inplace=True)
data['DS'].fillna(data['DS'].mean(), inplace=True)
data['DP'].fillna(data['DP'].mean(), inplace=True)
data['MLTV'].fillna(data['MLTV'].mean(), inplace=True)
data['Width'].fillna(data['Width'].mean(), inplace=True)
data['Tendency'].fillna(data['Tendency'].mean(), inplace=True)
data = data.dropna(subset=['NSP'])
```

```python
#here we handeled the missing values & checked for duplicates , no duplicate values and we
#C) Identify and correct any inconsistencies in data types (e.g., numerical values stored as strings)
data.dtypes
```

There are no inconsistencies in the data types, as all columns are correctly set to float64

```python
#D) Detect and treat outliers if necessary
sns.catplot(data=data, kind='box')
plt.figure(figsize= (15,6))
plt.show()
```

```python
outlier_cols = ['LB', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV', 'Tendency']

def remove_outliers_iqr(data, columns):
    for col in columns:
        Q1 = data[col].quantile(0.25)  # 25th percentile
        Q3 = data[col].quantile(0.75)  # 75th percentile
        IQR = Q3 - Q1  # Interquartile range

        #  outlier boundaries
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Detect outliers
        outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
        print(f"Column: {col} - Outliers detected: {len(outliers)}")

        # Remove outliers
        data = data[(data[col] >= lower_bound) & (data[col] <= upper_bound)]

    return data

# Apply outlier removal
df_cleaned = remove_outliers_iqr(data, outlier_cols)

# Save cleaned data (optional)
df_cleaned.to_csv('cleaned_data.csv', index=False)

# Show summary
print("Outliers removed. Cleaned data saved as 'cleaned_data.csv'.")
```

```python
# Load the cleaned dataset
df_cleaned = pd.read_csv('cleaned_data.csv')

# Columns where outliers were removed
outlier_cols = ['LB', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV', 'Tendency']

# Check if extreme values are still present
summary = df_cleaned[outlier_cols].describe()
print(summary)

# Check row count difference
original_df = pd.read_csv('Cardiotocographic.csv')  # Replace with the original file
print(f"Original Rows: {original_df.shape[0]}, Cleaned Rows: {df_cleaned.shape[0]}")
```
